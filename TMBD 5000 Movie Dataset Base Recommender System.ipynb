# Processing for Tags Column

Now we will parse all the jason data in the columns to build a tags column for content based recommendation system

```python
# After parsing jason it returns array in form of string but we need object that is why we are using literal_eval function
import ast 
def parse_jason(object):
    Temp = []
    for i in ast.literal_eval(object):
        Temp.append(i['name'])
    return Temp    
```

```python
data_Set_Movies['genres'] = data_Set_Movies.genres.apply(parse_jason)
```

```python
data_Set_Movies.head(5)
```

```python
#Applying the same function to keywords columns
data_Set_Movies.keywords.values
```

```python
#we only need name of keyword from the dictionary 
data_Set_Movies['keywords'] = data_Set_Movies.keywords.apply(parse_jason)
```

```python
data_Set_Movies.head(2)
```

```python
data_Set_Movies.cast.values
```

As the cast really affects the interests of Masses. We will fetch first three names of cast and we will improvise out parse_jason function 

```python
import ast
def improvised_parse_jason(object):
    Array = []
    count = 0
    for i in ast.literal_eval(object):
        if count < 3:
            Array.append(i['name'])
            count = count + 1
    return Array        
```

```python
data_Set_Movies['cast'] = data_Set_Movies.cast.apply(improvised_parse_jason)
```

```python
data_Set_Movies.head(3)
```

```python
data_Set_Movies.crew.values
```

Similarly from the crew column we only want name of director because movie enthusiast also has interest in the name of director.

```python
import ast
def parse_jason_crew(object):
    Array = []
    for i in ast.literal_eval(object):
        if i['job'] == 'Director':
            Array.append(i['name'])
    return Array 
```

```python
data_Set_Movies['crew'] = data_Set_Movies.crew.apply(parse_jason_crew)
```

```python
data_Set_Movies.head(5)
```

Now we will remove the space between the names of case and crew and other spaced keywords. The Reason is comprehensively explained in the project report.

```python
def space_remover(j):
    Array = []
    for i in j:
        Array.append(i.replace(" ",""))
    return Array    
```

```python
Temp_Array = ['crew', 'cast', 'keywords', 'genres']
for i in Temp_Array:
    data_Set_Movies[i] = data_Set_Movies[i].apply(space_remover)
```

```python
data_Set_Movies.head(2)
```

```python
#Overview column is also in string form. We will split it into Python List
#Split the overview column into array based on space delimiter
data_Set_Movies['overview'] = data_Set_Movies.overview.apply(lambda x: x.split(' '))
```

```python
data_Set_Movies.head(2)
```

```python
#Now merginf all the columns to make a separate tags column for language processing
data_Set_Movies['Movie Tags'] = data_Set_Movies['overview']+data_Set_Movies['genres']+data_Set_Movies['keywords']+data_Set_Movies['cast']+data_Set_Movies['crew']
```

```python
#Dropping all other columns becuase they are use less.
#And combining the Movie Tags Column Array in the form of string. Comprehensive reason is explained in report
data_Set_Movies['Movie Tags'] = data_Set_Movies['Movie Tags'].apply(lambda x: " ".join(x))
```

# Vectorization & Language Processing

```python
from sklearn.feature_extraction.text import CountVectorizer
#max_features meaning 5000 most frequent words will be selected
Vector = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectorize_results = Vector.fit_transform(Final_Data_Frame['Movie Tags']).toarray()
```

```python
#it has columns of 5000 each with the word that we have obtained from bag of words technique. Each Column represents one vector 
vectorize_results.shape
```

```python
# Feature Vectors
vectorize_results
```

```python
# All 5000 words from Movie Tags Column which are most frequently occuring
Vector.get_feature_names()
```

Now we will use the natural language tool kit to use stemming technique to avoid stopwords and almost similar words in the 5000 bag of words

```python
# Stemming technique to avoid duplication of same words
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
def Stemming_Technique(t):
    Array = []
    for i in t.split(' '):
        Array.append(stemmer.stem(i))
    return " ".join(Array)    
```

```python
Final_Data_Frame['Movie Tags'] = Final_Data_Frame['Movie Tags'].apply(Stemming_Technique)
```

```python
Final_Data_Frame['Movie Tags'].values
```

# Cosine Similarity

Now we will use the cosine similarity to check the cos angle between 5000 vectors and the ones with the closest cosine angle will be recommended by the system

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_Score = cosine_similarity(vectorize_results)
#Similarity score gives cosine angle of 1 for the same movie you searched for because of same vector
```

```python
# Recommendation Function
def Content_bases_Recommendation(movie_name):
    #Getting the index by using the name of Movie from the data set
    index = Final_Data_Frame[Final_Data_Frame['title'] == movie_name].index[0]
    #Key means sorting on the basis of second element of tuple which is cosine score
    cosine_distances = sorted(list(enumerate(similarity_Score[index])),reverse=True,key = lambda x: x[1])
    #Recommendations of first 5 movies
    for i in cosine_distances[1:6]:
        print(Final_Data_Frame.iloc[i[0]].title)
```

```python
Content_bases_Recommendation('Avatar')
```
