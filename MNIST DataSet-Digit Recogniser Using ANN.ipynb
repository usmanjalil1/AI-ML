# Description of DataSet

The MNIST dataset is a commonly used dataset in deep learning for image classification tasks. It comprises a set of 70,000 images of handwritten digits, with 60,000 images used for training and 10,000 images used for testing. Each image is 28 pixels by 28 pixels, resulting in 784 total pixels. Each pixel is represented by a value between 0 and 255, where 0 represents a white pixel and 255 represents a black pixel.

In addition to the images themselves, the dataset also includes corresponding labels that identify which digit is present in each image. There are 10 possible labels, one for each digit from 0 to 9.

To utilize this data for deep learning, we will first need to preprocess it by normalizing the pixel values to be between 0 and 1, converting the labels to one-hot encoded vectors, and splitting the data into training and testing sets.

We will then build an artificial neural network (ANN) using a deep learning framework such as TensorFlow or PyTorch. The ANN will have an input layer with 784 neurons, representing the 784 pixels in each image. This will be followed by one or more hidden layers, each containing a variable number of neurons. The final layer will have 10 neurons, representing the 10 possible output classes.

During training, the ANN will use backpropagation to adjust the weights and biases of the neurons in each layer in order to minimize the loss function. This will involve repeatedly feeding the training data through the network and comparing the predicted outputs to the actual labels.

Once the model has been trained, we can evaluate its performance on the testing set by calculating metrics such as accuracy, precision, recall, and F1 score. We can also use visualization techniques to better understand how the model is making predictions, such as by creating confusion matrices or plotting the activations of neurons in the hidden layers.

Overall, the MNIST dataset is a popular choice for deep learning projects due to its simplicity and ease of use. However, it still presents several challenges, such as overfitting, which must be addressed in order to build an effective model.

```python
X_train = X_train / 255
X_test = X_test / 255
```

Before using the MNIST dataset to train a machine learning model, it is common to preprocess the images by scaling the pixel values to be between 0 and 1. This is done by dividing each pixel value by 255, which is the maximum possible value for a pixel in the dataset.

The reason for this preprocessing step is to ensure that the pixel values are on a similar scale, which can help the machine learning model converge faster and perform better. If the pixel values were left in their original range of 0 to 255, the differences between pixel values could be too large, which could cause the machine learning model to learn features that are not relevant to the task at hand.

Therefore, dividing each pixel value by 255 is a common and effective way to preprocess the MNIST dataset before using it to train a machine learning model.

```python
X_train[0]
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

The reason for flattening the images before putting them into a machine learning model is that many machine learning algorithms, such as neural networks, require the input to be a 1D vector.

When the images are flattened, each pixel is represented as a single element in the input vector, and the entire image is represented as a long vector of pixel values. This vector can then be fed into a machine learning model as input.

Flattening the images also reduces the number of dimensions of the input data, making it easier to process and analyze the data. In addition, flattening the images can help reduce the computational cost of training the machine learning model, since many machine learning algorithms are more efficient at processing 1D vectors than 2D arrays.

Therefore, flattening the images before putting them into a machine learning model is a common and effective way to preprocess image data.

### Testing Accuracy with Different Layers in ANN  

```python
# Testing with only one input-output layer
DL_Model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])
```

```python
DL_Model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

DL_Model.fit(X_train_flattened, y_train, epochs=10)
```

```python
DL_Model.evaluate(X_test_flattened, y_test)
```

```python
import numpy as np
plt.matshow(X_test[1])
y_pred = DL_Model.predict(X_test_flattened)
np.argmax(y_pred[1])
```

```python
plt.matshow(X_test[80])
y_pred = DL_Model.predict(X_test_flattened)
np.argmax(y_pred[80])
```

```python
plt.matshow(X_test[1000])
y_pred = DL_Model.predict(X_test_flattened)
np.argmax(y_pred[1000])
```

It is observatory that we tried to test our model trained on one layer only and it gave us a almost 92% accuracy and it has detected all the 3 images right.

```python
import seaborn as sns
import tensorflow as tf
Predicted_labels = [np.argmax(i) for i in y_pred]
confusion_matrix = tf.math.confusion_matrix(labels=y_test,predictions=Predicted_labels)
plt.figure(figsize = (10,7))
sns.heatmap(confusion_matrix, annot=True,cmap='Blues')
plt.xlabel('Predicted Values')
plt.ylabel('Truth Values')
```

```python
#Lets make increase our accuracy with more dense and dropout layers
model2 = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(60, activation='sigmoid'),
    keras.layers.Dense(50, activation='sigmoid'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10, activation='sigmoid')
])

model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

plots = model2.fit(X_train_flattened, y_train, epochs=10, validation_data=(X_train_flattened, y_train))
plots
```

With 3 dense layers excluding output output layer and one dropout regularization layer we have achieved unbelievable accuracy of 99.18%.

During training, dropout randomly drops out a fraction of the neurons in a layer, meaning that the neurons do not participate in forward and backward passes, and their activations are set to zero. As a result, dropout reduces the co-adaptation of neurons and prevents the network from relying too heavily on any individual neuron or feature. This, in turn, improves the robustness of the network and makes it less sensitive to noise and irrelevant features.

Dropout can be added to any layer in a neural network, but it is most commonly used in the hidden layers, where it has been shown to be effective in improving the performance of a wide range of deep learning models.

Why we are using adam optimiser?
The Adam optimizer uses adaptive learning rates, which means that it adjusts the learning rate for each parameter based on the magnitude of the gradient and the previous update history. This helps the optimizer to converge faster and more reliably than other optimization algorithms such as vanilla SGD.

```python
plt.plot(plots.history['loss'], label='Training Loss')
plt.plot(plots.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss Curves')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(plots.history['accuracy'], label='Training Accuracy')
plt.plot(plots.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy Curves')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

From above two graphs it is clearly visible that how loss has drastically decrease while running 10 epochs on training and and validation dataset and how accuracy has increased. It shows model is being trained with high accuracy.

The categorical_crossentropy loss function is commonly used in multi-class classification tasks, where the output is a probability distribution over several classes. However, in cases where the output classes are mutually exclusive and represented as integers, we can use the sparse_categorical_crossentropy loss function. The sparse_categorical_crossentropy function is similar to categorical_crossentropy, but it is specifically designed to handle integer encoded labels rather than one-hot encoded labels.

In our case, since the MNIST dataset contains integer labels representing the digit classes, we used sparse_categorical_crossentropy as the loss function in our ANN. This allowed us to efficiently train the model to accurately classify the handwritten digits in the dataset.

Overall, the choice of optimization algorithm and loss function can significantly impact the performance of the neural network. By using ADAM as the optimizer and sparse_categorical_crossentropy as the loss function, we were able to achieve high accuracy on the MNIST dataset.
